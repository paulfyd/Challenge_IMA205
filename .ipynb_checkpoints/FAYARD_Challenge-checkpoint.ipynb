```python
def bilan_cnn_accuracy(history):
  ''' 
  Inputs: 
          history: a keras History object (obtained during the training)

  No Outputs

  Goal : Plot the evolution of the accuracy and of the loss function during the epochs          
  '''
  acc = history.history['accuracy']
  val_acc = history.history['val_accuracy']
  loss = history.history['loss']
  val_loss = history.history['val_loss']

  plt.figure(figsize=(8, 8))
  plt.subplot(2, 1, 1)
  plt.plot(acc, label='Training Accuracy')
  plt.plot(val_acc, label='Validation Accuracy')
  plt.legend(loc='upper right')
  plt.ylabel('Accuracy')
  plt.ylim([min(plt.ylim()),1])
  plt.title('Training and Validation Accuracy')

  plt.subplot(2, 1, 2)
  plt.plot(loss, label='Training Loss')
  plt.plot(val_loss, label='Validation Loss')
  plt.legend(loc='upper right')
  plt.ylabel('Cross Entropy')
  plt.ylim([0,1.0])
  plt.title('Training and Validation Loss')
  plt.xlabel('epoch')
  plt.show()

def bilan_cnn_f1(history):
  ''' 
  Inputs: 
          history: a keras History object (obtained in the training)

  No Outputs

  Goal : Plot the evolution of the f1_score and of the loss function during the epochs          
  '''
  acc = history.history['f1_m']
  val_acc = history.history['val_f1_m']
  loss = history.history['loss']
  val_loss = history.history['val_loss']

  plt.figure(figsize=(8, 8))
  plt.subplot(2, 1, 1)
  plt.plot(acc, label='Training F1')
  plt.plot(val_acc, label='Validation F1')
  plt.legend(loc='upper right')
  plt.ylabel('F1-score')
  plt.ylim([min(plt.ylim()),1])
  plt.title('Training and Validation F1-scores')

  plt.subplot(2, 1, 2)
  plt.plot(loss, label='Training Loss')
  plt.plot(val_loss, label='Validation Loss')
  plt.legend(loc='upper right')
  plt.ylabel('Cross Entropy')
  plt.ylim([0,1.0])
  plt.title('Training and Validation Loss')
  plt.xlabel('epoch')
  plt.show()
```

```python
#Recall, precision and F1-score, the metrics that interest us

def recall_m(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    recall = true_positives / (possible_positives + K.epsilon())
    return recall

def precision_m(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    return precision

def f1_m(y_true, y_pred):
    precision = precision_m(y_true, y_pred)
    recall = recall_m(y_true, y_pred)
    return 2*((precision*recall)/(precision+recall+K.epsilon()))
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    Goal : This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    (from TP2 IMA205)
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()

classes_name=["No person","Person"]
```

## 1.2 ) Haar-like features 

Haar-like features are digital image features used in object recognition. They owe their name to their intuitive similarity with Haar wavelets and were used in the first real-time face detector. A simple rectangular Haar-like feature can be defined as the difference of ***the sum of pixels of areas inside the rectangle***, which can be at any position and scale within the original image. 
The key advantage of a Haar-like feature over most other features is its ***calculation speed***. Due to the use of integral images, a Haar-like feature of any size can be calculated in constant time.

We can use the skimage.feature implementation to obtain this feature.

```python
def extract_feature_image(img, feature_type, feature_coord=None):
    """Extract the haar feature for the current image"""
    ii = integral_image(img)
    return haar_like_feature(ii, 0, 0, ii.shape[0], ii.shape[1],
                             feature_type=feature_type,
                             feature_coord=feature_coord)

feature_types = ['type-2-x', 'type-2-y']
```

It works only with black and white pictures, so we have to select one chanel of our images, let's say the first one.

```python
#An example
example_image=X[19]
example_image=resize(example_image,(25,25,3), mode='reflect')
example_image=example_image[:,:,0]
print("The shape of the features vector of each image is",extract_feature_image(example_image,feature_types).shape)
```

It's a failure: we didn't learn anything with these features since SVM classifies all the images of the test set in the "no humans" class. 

What's wrong? 

There are not enough features. 
We should select more features and then apply cascades of classifiers to classify.

Here is a method that would have been more efficient, but I didn't have time to implement it (AdaBoost is very long with 100000+ features). 

- Find the best features : To do this, we apply each and every feature on all the training images. For each feature, it finds the best threshold that will classify the images as positive and negative. Obviously, there will be errors or misclassifications. We select the features with the minimum error rate, which means they are the features that best classify the face and non-face images.

- The final classifier would be a weighted sum of these weak classifiers. It is called weak because it alone can’t classify the image, but together with others, it forms a strong classifier. We can select only 1000 features for example.

- After that, I must take an image take each 24×24 window, apply 1,000 features to it, and check if it is a human or not. That is too much time consuming but there is a solution : cascade classifier. In an image, most of the image region is non-human. So it is a better idea to have a simple method to verify that a window is not a human region. If it is not, discard it in a single shot. We don’t process it again. Instead, we only focus on the region where there can be a human. This way, we can find more time to check a possible human region.


- For this, the concept of a cascade of classifiers has been introduced. "Instead of applying all the 1,000 features to a window, we group the features into different stages of classifiers and apply one by one (normally first few stages will contain very few features). If a window fails in the first stage, discard it. We don’t consider the remaining features in it. If it passes, apply the second stage of features and continue the process. The window that passes all stages is a human region. How cool is the plan!!!"

## 1-C) Bag of words 

Here's another method I could have tested : the bag of words. 

Instead of working with features like Hog or Haar, I would have liked to work with SIFT. The problem with SIFT is that the number of features is different from one image to another. It's hard (impossible?) to train models on data that doesn't have the same number of features. 

Here is the solution: 

  - Extract all the feature points from each image with SIFT. To each point corresponds a vector of size 128. 
  - Gather all the vectors (128,) obtained. Cluster them, using k-means, with k=the desired vocabulary size. 
  - Create a Bag of words. For each image, take one by one the characteristic points detected by SIFT, and associate them to the corresponding cluster. This way, each image has only k clusters left. 
  - Apply a binary classifier from these vectors of size k 

SOURCE : 

https://arxiv.org/pdf/1505.00110.pdf (part 4.1) 

https://towardsdatascience.com/bag-of-visual-words-in-a-nutshell-9ceea97ce0fb


This method is costly in time and memory. Impossible to run it locally, because it crashes before having all the SIFT points. Also difficult to run it on Colab, because OpenCV's SIFT and SURF detectors don't seem to work. (cf the cell bellow)

To build our own CNN, we first have to choose its architecture.

Here is the implemenation of _small_ _ _model_ :

```python
small_model = Sequential()
small_model.add(Conv2D(input_shape=(224,224,3),filters=64,kernel_size=(3,3),padding="same", activation="relu"))
small_model.add(Conv2D(filters=64,kernel_size=(3,3),padding="same", activation="relu"))
small_model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
small_model.add(Conv2D(filters=128, kernel_size=(3,3), padding="same", activation="relu"))
small_model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
small_model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
small_model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
small_model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
small_model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
small_model.add(Flatten())
small_model.add(Dense(units=500,activation="relu"))
small_model.add(Dense(units=500,activation="relu"))
small_model.add(Dense(units=2, activation="softmax"))
```

### 2-A-2) A more complex architecture

So I decided to increase the number of layers. I chose a classical architecture that has already proven itself: that of VGG16. So I'm switching to 13 Conv layers. The new CNN should be able to learn more interesting features.

```python
def model_as_vgg():
  model = Sequential()
  model.add(Conv2D(input_shape=(224,224,3),filters=64,kernel_size=(3,3),padding="same", activation="relu"))
  model.add(Conv2D(filters=64,kernel_size=(3,3),padding="same", activation="relu"))
  model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
  model.add(Conv2D(filters=128, kernel_size=(3,3), padding="same", activation="relu"))
  model.add(Conv2D(filters=128, kernel_size=(3,3), padding="same", activation="relu"))
  model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
  model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
  model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
  model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
  model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
  model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
  model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
  model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
  model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
  model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
  model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
  model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
  model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
  model.add(Flatten())
  model.add(Dense(units=4096,activation="relu"))
  model.add(Dense(units=4096,activation="relu"))
  model.add(Dense(units=2, activation="softmax"))
  return model
```

```python
model=model_as_vgg()
print("Voici le second modèle :")
model.summary()
```

### 2-A-4) Adding regularization 

There are several ways of doing regulation in the CNNs. I tried adding layers of Dropout on a regular basis, but that didn't solve the problem. 

Another way to do it is to add a penalty to the loss of function, proportional to the norm 1 or 2 of the weights. 

I decided to add a penality on the weights of the last dense layer, to penalize a too complex model, that will give rise to overfitting.

```python
def model_as_vgg_regularized():
  model = Sequential()
  model.add(Conv2D(input_shape=(224,224,3),filters=64,kernel_size=(3,3),padding="same", activation="relu"))
  model.add(Conv2D(filters=64,kernel_size=(3,3),padding="same", activation="relu"))
  model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
  model.add(Conv2D(filters=128, kernel_size=(3,3), padding="same", activation="relu"))
  model.add(Conv2D(filters=128, kernel_size=(3,3), padding="same", activation="relu"))
  model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
  model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
  model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
  model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
  model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
  model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
  model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
  model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
  model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
  model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
  model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
  model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
  model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
  model.add(Flatten())
  model.add(Dense(units=4096,activation="relu"))
  model.add(Dense(units=4096,activation="relu"))
  model.add(Dense(units=2, activation="softmax",kernel_regularizer=l2(0.001)))
  return model
```

```python
model=model_as_vgg_regularized()
#Compilation
model.compile(loss='binary_crossentropy',optimizer=keras.optimizers.Adam(learning_rate=0.000007,decay=0.000007/30.0),metrics=['accuracy',f1_m])
```

```python
#Training 
t0=time.time()
history=model.fit(X_train, y_train, epochs=30, validation_data=(X_test,y_test),batch_size=20)
t1=time.time()
print("The training last "+str(t1-t0)+" s, ie ", str((t1-t0)/60.0)," mn.")
```

```python
#Bilan
bilan_cnn_f1(history)
y_pred_categorical=model.predict(X_test)
y_pred= np.argmax(y_pred_categorical, axis=1)
y_test0=inverse_to_categorical(y_test)

# Compute confusion matrix
cnf_matrix = confusion_matrix(y_test0, y_pred)
# Plot normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes= classes_name ,normalize=True)
plt.show()
```

The result is already a little better, even encouraging. We've reached 40% in the "there's one person" class for the first time.

## 2-B) Transfert learning 

---

Transfer learning is a machine learning method where a model developed for a task is reused as the starting point for a model on a second task. In our case, it is the method of taking a pre-trained model (the weights and parameters of a network that has been trained on a large dataset previously) and “fine-tuning” the model with our own dataset.

 Source : 


_https://towardsdatascience.com/kaggle-competition-image-classification-676dee6c0f23_

My model will do supervised learning as usual, but with some ***upstream work that will make it a lot easier***.

```python
 def format_image(img_file):
    """
    This function reads and formats an image so that it can be fed to the VGG16 network
    
    Parameters
    ----------
    img_file : image file name
    
    Returns
    -------
    img_out_vgg : the correctly formatted image for VGG16
    img : the image as read by the load_img function of keras.preprocessing.image
    """
    # read image
    img = load_img(img_file)
    # convert image to an array
    img_out = img_to_array(img)
    # preprocess the image to put in the correct format for use with the VGG16 network trained on imagenet
    img_out_vgg = vgg16.preprocess_input(img_out)
    # add a dimension at the beginning, coresponding to the batch dimension
    img_out_vgg = np.expand_dims(img_out_vgg, axis=0)
    return img_out_vgg, img

def unformat_image(img_in):
    """
    This function inverts the preprocessing applied to images for use in the VGG16 network
    
    Parameters
    ----------
    img_file : formatted image of shape (batch_size,m,n,3)
    
    Returns
    -------
    img_out : a m-by-n-by-3 array, representing an image that can be written to an image file
    """
    #get rid of batch dimension
    img_out=np.squeeze(img_in)
    #remove offsets added by the VGG16 preprocessing
    img_out[:, :, 0] += 103.939
    img_out[:, :, 1] += 116.779
    img_out[:, :, 2] += 123.68
    # invert the order of the colours : BGR -> RGB
    img_out = img_out[:, :, ::-1]
    #clamp image to the range [0,255] and cast to uint8
    img_out = np.clip(img_out, 0, 255).astype('uint8')
    return img_out
```

Then, we define the loss as the average response of the 15th channel of the 5th layer, second convolution. And we define the gradients of loss with respect to the image.
It enables us to maximizes the response to this particular chanel, and to see what it learns.

```python
#put the layers in an easy-to-reference dictionary
layer_dict = dict([ (layer.name,layer) for layer in cnnModel.layers])

layer_dict
```

```python
layer_out = layer_dict["block5_conv2"]

layer_out = layer_out.get_output_at(1)[:,:,:,9]
```

```python
#define loss
loss = K.mean(layer_out)
grads = K.gradients(loss, img_backend)[0]

# normalize the gradient
grads_normalised = grads/ (K.maximum(K.mean(K.abs(grads)), K.epsilon()))

get_loss_and_grads = K.function([img_backend],[loss,grads_normalised])
```

We the gradient descent, with our image as an iput, and see what have changed at the end. 

```python
img_in,img_visu = format_image('./im/IM_000301.jpg')
step = 0.1 # Gradient ascent step size
n_iterations = 1000  # Number of gradient ascent steps
for ii in range(0,n_iterations):
    loss_value,grads_value = get_loss_and_grads([img_in])
    img_in = img_in + grads_value * step
    if (ii%5==0):
        print(".", end='')
        img_out = unformat_image(np.copy(img_in))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4), sharex=True, sharey=True)
ax1.axis('off')
ax1.imshow(img_visu)
ax1.set_title('Input image')
ax2.axis('off')
ax2.imshow(img_out)
ax2.set_title('After Deep Dream')
plt.show()
```

### 2-C-2) Averaging the predictions

A good way to improve our mistakes is to ***average the predictions of each of the classifiers***. 
Everyone makes different mistakes. 

For example, for an image, it may be that the classifier is not sure of itself and returns [0.46,0.54]. So it will predict class 1. But the second classifier returns [0.97,0.03]. He is much more confident, and he is right because the real label is class 1. 
By taking the average of these two predictions, I will correct the error made by classifier 1, and maybe this same classifier 1 will correct other errors of classifier 2.

### 2-C-3) Stacking

So we tried to average out our predictions.A limitation of this approach is that each model contributes the same amount to the ensemble prediction, regardless of how well the model performed. A variation of this approach, called a weighted average ensemble, weighs the contribution of each ensemble member by the trust or expected performance of the model on a holdout dataset. This allows well-performing models to contribute more and less-well-performing models to contribute less. The weighted average ensemble provides an improvement over the model average ensemble.

A further generalization of this approach is ***replacing the linear weighted sum*** (e.g. linear regression) model used to combine the predictions of the sub-models ***with any learning algorithm***. This approach is called stacked generalization, or stacking for short.

Thus, we're going to use the output of the CNNs to train a new model.

Source : 

_https://machinelearningmastery.com/stacking-ensemble-for-deep-learning-neural-networks/_

As input for a new model, we will require examples with some number of features. Given that we have five models and each model makes two predictions per example, then we would have 10 (2 x 5) features for each example provided to the submodels. We can transform the [3350, 5, 2] shaped predictions from the sub-models into a [3350, 10] shaped array to be used to train a meta-learner using the reshape() NumPy function and flattening the final two dimensions. The stacked_dataset() function implements this step.

```python
#Train
y=[y2,y3,y4,y5]
stackX=y1
for yi in y:
  stackX = dstack((stackX, yi))
stackX = stackX.reshape((stackX.shape[0], stackX.shape[1]*stackX.shape[2]))

#Test
yt=[yt2,yt3,yt4,yt5]
stackX_t=yt1
for yti in y:
  stackX_t = dstack((stackX_t, yti))
stackX_t = stackX_t.reshape((stackX_t.shape[0], stackX_t.shape[1]*stackX_t.shape[2]))

#Final data (to be submitted)
yf=[yf2,yf3,yf4,yf5]
stackX_f=yf1
for yfi in yf:
  stackX_f = dstack((stackX_f, yfi))
stackX_f = stackX_f.reshape((stackX_f.shape[0], stackX_f.shape[1]*stackX_f.shape[2]))
```

```python
def main(f1=0):
  #x=0 or x=1
  #Choose 0 if you want the models that minimize the val loss and 1 those who maximize the val F1-score

  #X_exam contains the images that we want to predict
  test_images=load("test_images.npy")
  X_exam = test_images.astype('float32')
  X_exam /= 255.0

  #VGG
  vggmodel = VGG16(weights='imagenet', include_top=True)
  x = vggmodel.layers[-2].output
  predictions = Dense(2, activation="sigmoid")(x)
  vgg_model = Model(input = vggmodel.input, output = predictions)
  if f1==0:
    vgg_model.load_weights('weights_vgg(loss).hdf5')
  print("------------------------------------------------------")
  print("VGG loaded succesfully (1/5)")
  print("------------------------------------------------------")
  

  #Resnet
  resnet_model=keras.applications.resnet_v2.ResNet152V2(include_top=True, weights='imagenet')
  x = resnet_model.layers[-2].output
  predictions = Dense(2, activation="sigmoid")(x)
  resnet_model = Model(input = resnet_model.input, output = predictions)
  if f1==0:
    resnet_model.load_weights('weights_resnet(loss).hdf5')
  if f1!=0:
    resnet_model.load_weights('weights_resnet(f1).hdf5')
  print("------------------------------------------------------")
  print("Resnet loaded succesfully (2/5)")
  print("------------------------------------------------------")

  #Nasnet
  nasnet_model=keras.applications.NASNetMobile(include_top=True, weights='imagenet') 
  x = nasnet_model.layers[-2].output
  predictions = Dense(2, activation="sigmoid")(x)
  nasnet_model = Model(input = nasnet_model.input, output = predictions)
  if f1==0:
    nasnet_model.load_weights('weights_nasnet(loss).hdf5')
  if f1!=0:
    nasnet_model.load_weights('weights_nasnet(f1).hdf5')
  print("------------------------------------------------------")
  print("Nasnet loaded succesfully (3/5)")
  print("------------------------------------------------------")


  #Inception 
  inception_model=keras.applications.inception_resnet_v2.InceptionResNetV2(include_top=False, weights='imagenet', input_shape=(224,224,3))
  x = inception_model.layers[-2].output
  x = Flatten()(x)
  predictions = Dense(2, activation="sigmoid")(x)
  inception_model = Model(input = inception_model.input, output = predictions)
  if f1==0:
    inception_model.load_weights('weights_inception(loss).hdf5')
  if f1!=0:
    inception_model.load_weights('weights_inception(f1).hdf5')
  print("------------------------------------------------------")
  print("Inception loaded succesfully (4/5)")
  print("------------------------------------------------------")


  #Xcpetion
  xcpetion_model=keras.applications.xception.Xception(include_top=False, weights='imagenet', input_shape=(224,224,3))
  x = xcpetion_model.layers[-2].output
  x = Flatten()(x)
  predictions = Dense(2, activation="sigmoid")(x)
  xcpetion_model = Model(input = xcpetion_model.input, output = predictions)
  if f1==0:
    xcpetion_model.load_weights('weights_xcpetion(loss).hdf5')
  if f1!=0:
    xcpetion_model.load_weights('weights_xcpetion(loss).hdf5') 
  print("------------------------------------------------------")
  print("Xcpetion loaded succesfully (5/5)") 
  print("------------------------------------------------------")
  print("------------------------------------------------------")

  #Each yi is a prediction on the X_exam images that we must classify
  print("We are going to make the prediction for each model")
  y1=vgg_model.predict(X_exam)
  print("OK for model 1/5")
  y2=resnet_model.predict(X_exam)
  print("OK for model 2/5")
  y3=nasnet_model.predict(X_exam)
  print("OK for model 3/5")
  y4=inception_model.predict(X_exam)
  print("OK for model 4/5")
  y5=xcpetion_model.predict(X_exam)
  print("OK for model 5/5 !!!")
  print("------------------------------------------------------")

  #y_moy is the mean of the 5 prediction
  print("Now, let's average the predictions and return the new y_pred")
  y_moy=(y1+y2+y3+y4+y5)/5.0

  #Submit the results 
  y_moy=np.argmax(y_moy,axis=1)
  #Create the CSV in my Drive
  #submit_results(y_moy)

  return y_moy
```
